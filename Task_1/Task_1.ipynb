<h3> Prediction Using Supervised ML (Linear Regression)</h3>

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import metrics
```

```python
df=pd.read_csv("student_score.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
plt.scatter(df['Hours'],df['Scores'],)
```

```python
fig,axis=plt.subplots(1,2)

axis[0].boxplot(df['Hours'])
axis[0].set_xlabel("Hours")
axis[1].boxplot(df['Scores'])
axis[1].set_xlabel("Scores")
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
```

```python
x=np.asanyarray(df['Hours']).reshape(-1,1)
y=np.asanyarray(df['Scores']).reshape(-1,1)
```

```python
lr.fit(x,y)
```

```python
pre=np.array(9.25).reshape(-1,1)
# pre.append(9.25)
lr.predict(pre)
```

<h3> The Student Who study For <code>9.25 hrs/day</code> got <code>92 </code> score in exam</h3><br>
<h3>Hence we can say that as <u>Hours Of study</u> Increased the <u>score</u> increased</h3>
